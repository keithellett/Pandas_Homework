```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
#renaming a column
purchase_data = purchase_data.rename(columns= {'SN': 'Screen Name'})
```

## Player Count

```python
total_players = len(purchase_data['Screen Name'].unique())
total_players

total_players_summary = pd.DataFrame({'Total Players': [total_players]})
total_players_summary
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data['Item ID'].unique())
average_price = purchase_data['Price'].mean()
average_price = round(average_price,2)
total_purchases = purchase_data['Purchase ID'].count()
total_revenue = purchase_data['Price'].sum()
summary_table = pd.DataFrame({'Unique Item': unique_items,
                              'Average Price in $':[average_price],
                              'Total Purchases': total_purchases,
                              'Total Revenue in $': [total_revenue]
})
summary_table
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
print(purchase_data['Age'].max())
print(purchase_data['Age'].min())
```

```python
#creating bins  broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
bins = [0, 9.9 , 14, 18, 22, 26, 30, 34, 38, 42, 46]
age_groups = ['<10', '10-14','14-18','18-22','22-26','26-30','30-34','34-38','38-42','42-46']
```

```python
pd.cut(purchase_data['Age'], bins, labels = age_groups).head()
```

```python
purchase_data['Age Group'] = pd.cut(purchase_data['Age'], bins, labels = age_groups)
purchase_data.head()
```

```python
age_df = purchase_data.drop_duplicates(subset='Screen Name')['Age Group'].value_counts()
age_df
age_df= pd.DataFrame(age_df)
age_df.head()
age_df['% of Total'] = round(age_df['Age Group']/total_players, 4)*100
age_df.head(10)
```

```python
#purchase_data['Age Group'].value_counts
#age_df['% of Total'] = round(purchase_data['Age Group']/total_players, 2)*100
age_grouped = purchase_data.groupby('Age Group')
print(age_grouped['Purchase ID'].count())




```

```python

#create variables
total_purch_by_age = age_grouped['Purchase ID'].count()
avg_purch_price_by_age = round(age_grouped['Price'].mean(), 2)
total_purch_price_by_age = age_grouped['Price'].sum()

age_summary = pd.DataFrame({'Purchase Count': total_purch_by_age,
                              'Avg Purch Price in $': avg_purch_price_by_age,
                              'Total Purch Amt in $': total_purch_price_by_age})
                              
age_summary['Avg TTL Purchase/Person in $'] = round(age_summary['Total Purch Amt in $']/ age_df['Age Group'],2)
age_summary.head(20)

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting

```python
screen_name_grouped = purchase_data.groupby(['Screen Name'])
screen_name_grouped.count().head()
```

```python

total_spent = screen_name_grouped['Price'].sum()
sn_purch_count = screen_name_grouped['Purchase ID'].count()

custy_summary = pd.DataFrame({'Total Spent': total_spent,
                              '# of Purch': sn_purch_count})

                              
custy_summary['Avg Purch Price in $'] = round(custy_summary['Total Spent'] / custy_summary['# of Purch'],2)

custy_summary.sort_values(by='Total Spent', ascending=False).head()
```
